```python
import numpy as np

class Perceptron(object):
    """Perceptron classifier.
    
    Parameters
    -----------
    eta : float
        Learning rate [0,1]
    n_iter : int
        Passes over the training dataset.
    random_state : int
        Random number generator seed for random weight
        initialization.
    
    Attributes
    -----------
    w_ : 1d-array
        Weights after fitting.
    errors_ : list
        Number of misclassifications (updates) in each epoch.
    """
    
    def __init__(self, eta=0.01, n_iter=50, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state
        
    def fit(self, X, y):
        """Fit training data.
        
        Parameters
        -----------
        X: {array-like}, shape = [n_examples, n_features]
            Training vectors, where n_Examples is the number of examples and n_features is the number of features.
        y: array-like, shape = [n_examples]
            Target values.
        
        Returns
        -----------
        self : object
        
        """
        rgen = np.random.RandomState(self.random_state) # random number generator
        self.w_ = rgen.normal(loc=0.0, scale=0.01, size=1 + X.shape[1]) # initialize weights
        self.errors_ = []
        
        for _ in range(self.n_iter):
            errors = 0
            for xi, target in zip(X, y):
                update = self.eta * (target - self.predict(xi)) # weight updating
                self.w_[1:] += update * xi # update each feature
                self.w_[0] += update # update bias
                errors += int(update != 0.0) # number of misclassifications during each epoch
            self.errors_.append(errors)
        return self # trained model
    
    def net_input(self, X):
        """Calculate net input"""
        return np.dot(X,self.w_[1:])+self.w_[0] # output z = wTx + bias - numpy is vectorized
    
    def predict(self, X):
        """Return class label after unit step"""
        return np.where(self.net_input(X) >= 0.0, 1, -1) # boolean function
```
